```python
import pandas_datareader as pdr
import pandas as pd
from datetime import datetime
```

```python
import yfinance
## Tesla Dataset
df_tesla = yfinance.download (tickers = "TSLA", start = "2017-02-27", 
                              end = "2022-03-13", interval = "1D", group_by = 'ticker', auto_adjust = True, threads = True)
```

```python
df_tesla.head()
```

```python
## Simple Moving average
```

```python
df_tesla['Open'].plot(figsize=(12,4))
```

```python
df_tesla['Open:10days rolling']=df_tesla['Open'].rolling(window=10,min_periods=1).mean()
```

```python
df_tesla[['Open','Open:10days rolling']].plot(figsize=(15,6))
```

```python
## Expanding

## Cumulative Moving average
df_tesla['Open'].expanding().mean().plot(figsize=(10,5))
```

```python
#Exponential moving average
```

```python
df_tesla['EMA_0.1']=df_tesla['Open'].ewm(alpha=0.1,adjust=False).mean()
```

```python
df_tesla[['Open','EMA_0.1']].plot(figsize=(15,6))
```
